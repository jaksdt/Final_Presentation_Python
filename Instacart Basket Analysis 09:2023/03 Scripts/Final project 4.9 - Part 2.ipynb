## Step 3 - Creating histogram to show stakeholders descriptive findings about sales

```python
hist3 = instacart_sub['order_hour_of_day'].plot.hist(bins = 75)
```

### Histogram shows the frequency of shoppers through out the day. It starts getting really busy around 09:00 and continues until 16:00. It slowly dies down and from 00:00 - 06:00 it's dead quiet.

## Step 4 - Showing marketing team the most loyal customers

```python
bar2 = instacart_sub['loyalty_flag'].value_counts().plot.bar()
```

## Step 5 - Checking to see if there's a difference in expenditure in the prices column throughout the day.

```python
line2 = sns.lineplot(data = instacart_sub, x = 'order_hour_of_day',y = 'prices')
```

### Around 10:00, which is peak hour, people tend to spend the least amount of money per item. The rest of the day is fairly even with a coupl of small dips and spikes through out.

## Step 6 - Exploring connection between age and family situation for targeted marketing campaigns.

```python
# Line chart to show marital status
line4 = sns.lineplot(data = instacart_sub, x = 'age',y = 'marital_status')
```

```python
# Double-checking marital status
bar3 = instacart_sub['marital_status'].value_counts().plot.bar()
```

```python
# Line chart to show number of dependants in each family
line3 = sns.lineplot(data = instacart_sub, x = 'age',y = 'number_of_dependants')
```

### The vast majority of customers are maried and most customers have at least one dependant.

## Step 7 - Exploring connection between age and spending power with scatterplot.

```python
scatter = sns.scatterplot(x = 'age', y = 'income',data = instacart_sub)
```

### The older people get, the more money they make.
